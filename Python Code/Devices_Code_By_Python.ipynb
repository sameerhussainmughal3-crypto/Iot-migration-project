```python
import json
import time
import random
import threading
from datetime import datetime, timezone
from awscrt import mqtt, auth
from awsiot import mqtt_connection_builder

IOT_ENDPOINT = "a3mjjixaoa2wyh-ats.iot.us-east-1.amazonaws.com"
REGION = "us-east-1"
TOPIC = "iot/geolocation"
NUM_DEVICES = 5

BASE_LAT = 51.5030
BASE_LON = 0.0032

def build_connection(client_id):
    credentials_provider = auth.AwsCredentialsProvider.new_default_chain()
    return mqtt_connection_builder.websockets_with_default_aws_signing(
        endpoint=IOT_ENDPOINT,
        region=REGION,
        credentials_provider=credentials_provider,
        client_id=client_id,
        clean_session=True,
        keep_alive_secs=30
    )

def simulate_device(device_id):
    client_id = f"simulator-device-{device_id}"
    conn = build_connection(client_id)
    connect_future = conn.connect()
    connect_future.result()
    print(f"[device_{device_id}] connected")

    while True:
        payload = {
            "device_id": f"device_{device_id}",
            "latitude": BASE_LAT + random.uniform(-0.01, 0.01),
            "longitude": BASE_LON + random.uniform(-0.01, 0.01),
            "timestamp": datetime.now(timezone.utc).isoformat()
        }
        conn.publish(
            topic=TOPIC,
            payload=json.dumps(payload),
            qos=mqtt.QoS.AT_LEAST_ONCE
        )
        print(f"[device_{device_id}] sent: {payload}")
        time.sleep(5)

threads = []
for i in range(1, NUM_DEVICES + 1):
    t = threading.Thread(target=simulate_device, args=(i,), daemon=True)
    t.start()
    threads.append(t)
    time.sleep(1)

print("5 devices simulate ho rahe hain...")
```
